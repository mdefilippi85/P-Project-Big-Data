#Project description: 
The purpose of this work is to create an interactive map diplaying some information about the Earth's surface movement.
This can be made possible by means of long temporal series of Satellite images properly processed. 
The resulting map displays georeferenced points representing the average displacement rate [mm/year] of each target at ground. 
This map is colored by means of the rainbow color scale (as default) which is a diverging scale representing both negative and positive values: 

*   positive values of velocity mean uplift (up direction)
*   negative values of velocity mean subsidence (down direction) 

**Input Data**: point shape file (.shp, .shx, .dbf) contaiing information about the average displacement rate for each position at ground.                                         
**Output**: Average displacement rate map. For each point are shown the lat/long coordinate, average velocity[mm/yr] and coherence. Coherence is an index of quality ranging from 0 to 1 (e.g. correlation).                            
**Libraries** **used**: geopandas, google.colab, plotly.express.  
**Terminology**: Average Displacement rate is equivalent to average velocity.  
**New developments**: Improve interaction by adding displacement time series for each georeferenced point.

```
df_regioni = gpd.read_file(filename="/content/sentinel_asc_mosul_PS_75_0.shp")
df_regioni.head()
```

```
text = 'Displacement rate [mm/yr]' #@param {type:"string"}
dropdownMeasure = 'velocity' #@param ["velocity", "coherence", "MuSigma"]
colorMap = 'spectral' #@param ["rainbow", "greys", "spectral","thermal"]
lowerLimit =  -20#@param {type:"number"} 
upperLimit =  20#@param {type:"number"} 


fig = px.scatter_mapbox(df_regioni, lat='Lat', lon='Lon', hover_data=[dropdownMeasure, "coherence" ],  zoom=10, height=600,  color=dropdownMeasure, color_continuous_scale = colorMap, range_color= [lowerLimit,upperLimit]) 
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
#fig.add_scatter(lista_date[18:length-1],row[18:length-1])

fig.show()
```
