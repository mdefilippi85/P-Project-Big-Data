```python
import dash
from jupyter_plotly_dash import JupyterDash
import dash_core_components as dcc
import dash_html_components as html
import pandas as pd
import plotly.express as px
import geopandas as gpd 
from dash.dependencies import Input, Output 
import json 
import plotly.graph_objects as go 
import urllib

external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']

app = dash.Dash(name=__name__, external_stylesheets=external_stylesheets)

df_asc = gpd.read_file(filename="Lavoro/Processing/PS Datasets/London/London_small_asc_PS_processing/geocoding/London_small_asc_PS_85_0.shp")

#df_desc = gpd.read_file(filename="Lavoro/Processing/PS Datasets/London/London_small_desc_PS_processing/geocoding/London_small_desc_PS_85_0.shp")

#df_horiz = gpd.read_file(filename="Lavoro/Processing/PS Datasets/London/disp decomp/PS_London_85_decomposed_east.shp")

#df_vert = gpd.read_file(filename="Lavoro/Processing/PS Datasets/London/disp decomp/PS_London_85_decomposed_vertical.shp")

#df_regioni.head()

lista_date=list(df_asc.columns) 
length=len(lista_date)
lista_date[18:length-1]

field_to_show =["velocity", "coherence", "MuSigma"]

type_disp_to_show = ["Ascending", "Descending", "Horizontal", "Vertical"]
disp_dict={"Ascending":df_asc, "Descending":df_asc, "Horizontal":df_asc, "Vertical":df_asc}

label_field_to_show ={"Ascending":"velocity [mm/yr]","Descending":"velocity [mm/yr]","Horizontal":"velocity [mm/yr]","Vertical":"velocity [mm/yr]", "coherence":"coherence", "MuSigma":"MuSigma"}
colour_scales=["rainbow", "gray"] 

text = 'Displacement rate [mm/yr]' #@param {type:"string"}
dropdownMeasure = 'velocity' #@param ["velocity", "coherence", "MuSigma"]
colorMap = 'rainbow' #@param ["rainbow", "greys", "spectral","thermal"]
lowerLimit =  -20#@param {type:"number"} 
upperLimit =  20#@param {type:"number"} 

lista_date=list(df_asc.columns) 
length=len(lista_date)
lista_date[18:length-1]
    
#fig = px.scatter_mapbox(df_asc, lat='Lat', lon='Lon', hover_data=[dropdownMeasure, "coherence" ],  zoom=10, height=600,  color=dropdownMeasure, color_continuous_scale = colorMap, range_color= [lowerLimit,upperLimit]) 
#fig.update_layout(mapbox_style="open-street-map")
#fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

app.layout = html.Div([
    html.Div([
        
       
        html.Div([
            html.Label('Geometry'),
            dcc.Dropdown(
                id='disp_to_show_id',
                options=[{'label': i, 'value': i} for i in type_disp_to_show],
                value='Ascending'
            ),

        ],
        style={'height': '70%', 'width': '49%', 'display': 'inline-block'}),
        
        html.Div([
            html.Label('Layer Type'),
            dcc.Dropdown(
                id='field_to_show_id',
                options=[{'label': i, 'value': i} for i in field_to_show],
                value='velocity'
            ),

        ],
        style={'height': '70%', 'width': '49%', 'float': 'right', 'display': 'inline-block'}),
        
        html.Div([
            html.Label('Colour scale'),
            dcc.Dropdown(
                id='colour_scale_id',
                options=[{'label': i, 'value': i} for i in colour_scales],
                value='rainbow'
            ),
             
        ], style={'width': '49%', 'float': 'right', 'display': 'inline-block'}),
        
        html.Div([
           
            html.Label('Lower Limit'),
            dcc.Input( id='lower_limit_id', value=-20.0, type='number'),            
        ], 
        style={'width': '10%', 'display': 'inline-block'}), 
        
        html.Div([
              
             html.Label('Upper Limit'),
             dcc.Input(id='upper_limit_id', value=20.0, type='number'), 
         ], 
        style={'width': '10%', 'display': 'inline-block'}), 
            
    ], style={
        'borderBottom': 'thin lightgrey solid',
        'backgroundColor': 'rgb(250, 250, 250)',
        'padding': '10px 5px'
    }), 
    
    html.Div([
        dcc.Graph(
            id='map_velocity_id' ,
            hoverData = {'points': [{'pointNumber': '5'}]},
            clickData = {'points': [{'pointNumber': '5'}]}
        )
    ]),
    
     html.Div([
        dcc.Graph(id='time_series_id')]), 
         
     html.Div([
            
                #html.Button('Export To CSV', id='btn-nclicks-1', n_clicks=0),
             #     html.P(id='save-button-hidden', style={'display':'none'})]),
                  html.A(
                        'Export Time Series CSV',
                        id='download-link',
                        download="PS Time Series.csv",
                        href="",
                        target="_blank"
              )], style={'marginLeft': '4%'}),
              #html.Div(id="save-button-hidden",children="Press button to save changes")]), 
    
     html.Div([
            dcc.Markdown("""
                **Hover Data**

                Mouse over values in the graph.
            """),
            html.Pre(id='hover-data')
    ], className='three columns'),
    
    html.Div([
            dcc.Markdown("""
                **Click Data**

                Click on points in the graph.
            """),
            html.Pre(id='click-data')
        ], className='three columns')
    
   
  ])

# Mostra i field dell'oggetto hover_data 
@app.callback(
    Output('hover-data', 'children'),
    [Input('map_velocity_id', 'hoverData')])
def display_hover_data(hoverData):
    return json.dumps(hoverData, indent=2)

# Mostra i field dell'oggetto click data 
@app.callback(
    Output('click-data', 'children'),
    [Input('map_velocity_id', 'clickData')])
def display_click_data(clickData):
    return json.dumps(clickData, indent=2)

#callback: CREATE MAP 

@app.callback(
    Output('map_velocity_id', 'figure'),
    [Input('colour_scale_id', 'value'),
     Input('disp_to_show_id', 'value'), 
     Input('field_to_show_id', 'value'), 
     Input('lower_limit_id', 'value'),  
     Input('upper_limit_id', 'value')]) 

def update_graph(colour_scale_value, disp_to_show_value, field_to_show_value, lower_limit_value, upper_limit_value):

    fig = px.scatter_mapbox(disp_dict[disp_to_show_value], lat='Lat', lon='Lon', hover_data=[field_to_show_value, "coherence" ],  zoom=10, height=800,  color=field_to_show_value, color_continuous_scale = colour_scale_value, range_color= [lower_limit_value,upper_limit_value]) 
    fig.update_layout(coloraxis_colorbar=dict(title=label_field_to_show[disp_to_show_value])) 

    fig.update_layout(mapbox_style="open-street-map")
    fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

    return fig

#CREATE TIME SERIES 
def create_time_series(row):


    fig = go.Figure([go.Scatter(x=lista_date[17:length-1], y=row[17:length-1])])

    fig.update_layout(
        xaxis_title="Acquisition dates",
        yaxis_title="Displacement [mm]",
        title={
            'text': "Displacement Time Series LOS",
            'y':0.9,
            'x':0.5,
            'xanchor': 'center',
            'yanchor': 'top'}) 

    return fig

@app.callback(
    dash.dependencies.Output('time_series_id', 'figure'),
    [dash.dependencies.Input('map_velocity_id', 'clickData')])

def update_timeseries(clickData):
   
    id_record = int(clickData['points'][0]['pointNumber']) 

    row=df_asc.iloc[id_record]
   # title = '<b>{}</b><br>{}'.format("Displacement Time-Series", "Acquisition Dates")
    return create_time_series(row)

#Export to CSV: 

@app.callback(
    Output('download-link', 'href'),
    [Input('map_velocity_id', 'clickData'),
     Input('disp_to_show_id', 'value')]) 
   
def export_to_csv(clickData, disp_to_show_value):

    id_record = int(clickData['points'][0]['pointNumber']) 

    row=disp_dict[disp_to_show_value].iloc[id_record]
   # msg=''
   # csv_string=''
   # if n_clicks > 0:
    csv_string = row.to_csv(index=False, encoding='utf-8')
    csv_string = "data:text/csv;charset=utf-8," + urllib.parse.quote(csv_string)
        #row.to_csv('D:/Lavoro/time_Series.csv', index=False, encoding='utf-8')
   # msg = 'Data Submitted'
    
   # html.Div(msg)
    return csv_string
   



if __name__ == "__main__":
        app.run_server(port=7000, host='127.0.0.1', debug=True, use_reloader=False)
        
```
